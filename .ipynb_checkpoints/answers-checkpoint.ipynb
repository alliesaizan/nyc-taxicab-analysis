# NYC Taxicab Analysis

The data covers green taxicab trips during March 2020. In the analysis below, parts 1-4 consist of data exploration, parts 5-6 consist of tip percentage modelling, and part 7 provides geospatial visualizations about taxi dropoffs.

## Part 1

The code chunk below imports trip data from the NYC TLC, imports demographic information by taxi zone, and merges the demographic information to the taxi ride dataset. I assume we will want to include demographic information about the pickup (PU) and dropoff (DO) zones, so I merge the demographic information on the main dataset twice. The merged dataset representing trips in March 2020 has 223,496 rows. 

```python
# Package imports and notebook settings
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import geopandas
import geoplot
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score, mean_absolute_error

import warnings
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    
%matplotlib inline
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 100)
```

```python
# Read in taxicab and demographic data
taxi_rides = pd.read_parquet("https://s3.amazonaws.com/nyc-tlc/trip+data/green_tripdata_2020-03.parquet")
demographic_info = pd.read_csv("data/raw/taxi_zone_demographics.csv")
demographic_info.drop(columns = ["Unnamed: 0"], axis = 1, inplace = True)

# Merge taxi ride and demographic data./
    # We will merge twice, first to include the demographic information of the taximeter engage zone, 
    # and second to include the demographic information of the taximeter disengage zone.
departure_zone = demographic_info.copy()
departure_zone.columns = [col + "_PU" for col in departure_zone.columns]
departure_zone.reset_index(drop = True, inplace = True)

arrival_zone = demographic_info.copy()
arrival_zone.columns = [col + "_DO" for col in arrival_zone.columns]
arrival_zone.reset_index(drop = True, inplace = True)

taxi_rides_merged = pd.merge(left = taxi_rides, right = departure_zone, how = "left", left_on = "PULocationID", right_on = "Taxi zone_PU")
taxi_rides_merged = pd.merge(left = taxi_rides_merged, right = arrival_zone, how = "left", left_on = "DOLocationID", right_on = "Taxi zone_DO")

# Print out the rows and columns of each dataframe
taxi_rides_merged.shape
```

## Part 5

The code blocks below train a gradient boosting regression model to predict tip percentage. We do not have demographic information for all the taxi zones, so I drop rows with missing demographic information.

```python
# Create a tip percentage variable and drop the correlated columns
taxi_rides_merged["tip_percentage"] = taxi_rides_merged["tip_amount"] / taxi_rides_merged["fare_amount"]
```

```python
taxi_rides_merged.ehail_fee.unique() # E-Hail fee is none throughout, so drop it. Also drop the NTA Codes and datetime fields

# Drop the zone columns
taxi_rides_ml = taxi_rides_merged.drop(columns = ["Taxi zone_PU", "Taxi zone_DO", "tip_amount", "fare_amount"], axis = 1)

# One-hot encode the type, flag, and ID columns
taxi_rides_ml = pd.get_dummies(taxi_rides_ml, columns = ["store_and_fwd_flag", "VendorID", "RatecodeID", "payment_type", "trip_type"], drop_first = True)

# Remove any remaining object/datetime columns
taxi_rides_ml = taxi_rides_ml.select_dtypes(exclude = ["object", "datetime"])
taxi_rides_ml.head()
```

```python
# Run and evaluate a very basic gradient boosting regressor

# Drop missing values, leaving roughly half the records. Because we explicitly want to investigate demographic information in the next section, I do not want to impute demographic 
    # information and potentially introduce misleading data to the dataset.
taxi_rides_ml = taxi_rides_ml.dropna()
taxi_rides_ml.shape

taxi_rides_ml = taxi_rides_ml.sample(frac = 1) # shuffle the dataset to ensure the algorithm is not merely memorizing the data
X, y = taxi_rides_ml.drop(columns = ["tip_percentage"], axis = 1), taxi_rides_ml[["tip_percentage"]]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, random_state=123)

rf = GradientBoostingRegressor(n_estimators = 250, random_state=123)

rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

print(f"The baseline model's R2 score is {r2_score(y_test, y_pred)} and its mean absolute error is {mean_absolute_error(y_test, y_pred)}.")
```

## Part 6

The next code chunks evaluate the results by pickup and dropoff location demographics. I look first at the distribution of the black and nonwhite population in pickup and dropoff zones. Then, I set a threshold where the population rates are greater than the median. I analyze the average tip percentage and model error above and below this threshold. The results appear to show that predicted tip percentages in general are lower in zones with Black and Hispanic population rates that are higher than the median. Additionally, the mean absolute error is higher for high-Black population pickup zones. 

```python
# Set up thresholds (at equal to or greater than the median)
comparison_dataset = X_test.copy()

comparison_dataset["BlackPU_Threshold"] = np.where(comparison_dataset["BlackNonHisp_PU"] >= comparison_dataset["BlackNonHisp_PU"].median(), 1, 0)
comparison_dataset["HispanicPU_Threshold"] = np.where(comparison_dataset["Hispanic_PU"] >= comparison_dataset["Hispanic_PU"].median(), 1, 0)
comparison_dataset["HispanicDO_Threshold"] = np.where(comparison_dataset["Hispanic_DO"] >= comparison_dataset["Hispanic_DO"].median(), 1, 0)
comparison_dataset["BlackDO_Threshold"] = np.where(comparison_dataset["BlackNonHisp_DO"] >= comparison_dataset["BlackNonHisp_DO"].median(), 1, 0)

# Merge predictions and calculate MAE for each row
taxi_rides_ml_merged = pd.concat([comparison_dataset.reset_index(drop=True), y_test.reset_index(drop = True), pd.DataFrame({"predicted_tip_pct": y_pred})], axis = 1)

taxi_rides_ml_merged["mae"] = taxi_rides_ml_merged.apply(lambda row: mean_absolute_error([row.tip_percentage], [row.predicted_tip_pct] ), axis = 1)
```

```python
# Show the Mean Absolute Error for the zones where the Black and Hispanic population rates are greater than the 75th percentile.

mae_bPU = mean_absolute_error(y_true = taxi_rides_ml_merged.loc[taxi_rides_ml_merged["BlackPU_Threshold"] == 1, "tip_percentage"], y_pred = taxi_rides_ml_merged.loc[taxi_rides_ml_merged["BlackPU_Threshold"] == 1, "predicted_tip_pct"] )
mae_bDO = mean_absolute_error(y_true = taxi_rides_ml_merged.loc[taxi_rides_ml_merged["BlackDO_Threshold"] == 1, "tip_percentage"], y_pred = taxi_rides_ml_merged.loc[taxi_rides_ml_merged["BlackDO_Threshold"] == 1, "predicted_tip_pct"] )

print(f"The MAE for high-Black population pickup zones is {np.round(mae_bPU, 3)} and for dropoff zones is {np.round(mae_bDO, 3)}.")

mae_hPU = mean_absolute_error(y_true = taxi_rides_ml_merged.loc[taxi_rides_ml_merged["HispanicPU_Threshold"] == 1, "tip_percentage"], y_pred = taxi_rides_ml_merged.loc[taxi_rides_ml_merged["HispanicPU_Threshold"] == 1, "predicted_tip_pct"] )
mae_hDO = mean_absolute_error(y_true = taxi_rides_ml_merged.loc[taxi_rides_ml_merged["HispanicDO_Threshold"] == 1, "tip_percentage"], y_pred = taxi_rides_ml_merged.loc[taxi_rides_ml_merged["HispanicDO_Threshold"] == 1, "predicted_tip_pct"] )

print(f"The MAE for high-Hispanic population pickup zones is {np.round(mae_hPU, 3)} and for dropoff zones is {np.round(mae_hDO, 3)}.")
```

```python
# Show the difference in predicted fares across subgroups

tip_bPU = taxi_rides_ml_merged.loc[taxi_rides_ml_merged["BlackPU_Threshold"] == 1, "predicted_tip_pct"].mean()
tip_bDO = taxi_rides_ml_merged.loc[taxi_rides_ml_merged["BlackDO_Threshold"] == 1, "predicted_tip_pct"].mean()

print(f"The average predicted tip percentage for high-Black population pickup zones is {np.round(tip_bPU, 3)} and for dropoff zones is {np.round(tip_bDO, 3)}. In contrast, the average predicted tip percentage across all zones is {np.round(y_pred.mean(), 3)}.")

tip_hPU = taxi_rides_ml_merged.loc[taxi_rides_ml_merged["HispanicPU_Threshold"] == 1, "predicted_tip_pct"].mean()
tip_hDO = taxi_rides_ml_merged.loc[taxi_rides_ml_merged["HispanicDO_Threshold"] == 1, "predicted_tip_pct"].mean()

print(f"The average predicted tip percentage for high-Hispanic population pickup zones is {np.round(tip_hPU, 3)} and for dropoff zones is {np.round(tip_hDO, 3)}. In contrast, the average predicted tip percentage across all zones is {np.round(y_pred.mean(), 3)}.")
```

## Part 7

In the cells below, I use the geopandas package to visualize dropoffs across NYC. As should be expected, most dropoffs occur in Manhattan, and Queens near LaGuardia is the second-most common dropoff location. I was suprised to see that not as many dropoffs occurred in the JFK area as near LaGuardia. I also looked at the number of trips after March 11, 2020, as this is the day that the WHO declared the COVID-19 pandemic. I expect to see trips decline after this date. As anticipated, the maximum number of dropoffs declined by a significant amount.

```python
taxi_rides_count = taxi_rides_merged["DOLocationID"].value_counts().reset_index()
taxi_rides_count.columns = ["DOLocationID", "count"]
```

```python
# Import taxi zone shapefile
geo_data = geopandas.read_file("data/geo/taxi_zones.shp")
geo_data.head()
```

```python
# Merge to the taxi data dataset
taxi_rides_geo = pd.merge(left = geo_data, right = taxi_rides_count, how = "left", left_on = "LocationID", right_on = "DOLocationID")
taxi_rides_geo.shape
```

```python
fig, ax = plt.subplots(1, figsize=(20, 8))
ax.axis('off')
ax.set_title("Dropoffs by taxi zone", fontdict={'fontsize': '25', 'fontweight': '3'})
taxi_rides_geo.plot(column="count", ax=ax, cmap = "Blues")

sm = plt.cm.ScalarMappable(norm=plt.Normalize(vmin=taxi_rides_geo["count"].min(), vmax=taxi_rides_geo["count"].max()), cmap="Blues")
sm._A = []
cbaxes = fig.add_axes([0.15, 0.25, 0.01, 0.4])
cbar = fig.colorbar(sm, cax=cbaxes)
```

```python
# Extract data from after the start of the pandemic
taxi_rides_merged["lpep_dropoff_date"] = taxi_rides_merged["lpep_dropoff_datetime"].apply(lambda x: x.day)
after_pandemic_counts = taxi_rides_merged.loc[taxi_rides_merged["lpep_dropoff_date"] >= 12,"DOLocationID"].value_counts().reset_index()
after_pandemic_counts.columns = ["DOLocationID", "count"]
```

```python
# Plot
taxi_rides_geo2 = pd.merge(left = geo_data, right = after_pandemic_counts, how = "left", left_on = "LocationID", right_on = "DOLocationID")

fig, ax = plt.subplots(1, figsize=(20, 8))
ax.axis('off')
ax.set_title("Dropoffs by taxi zone (after March 11, 2020)", fontdict={'fontsize': '25', 'fontweight': '3'})
taxi_rides_geo2.plot(column="count", ax=ax, cmap = "Blues")

sm = plt.cm.ScalarMappable(norm=plt.Normalize(vmin=taxi_rides_geo2["count"].min(), vmax=taxi_rides_geo2["count"].max()), cmap="Blues")
sm._A = []
cbaxes = fig.add_axes([0.15, 0.25, 0.01, 0.4])
cbar = fig.colorbar(sm, cax=cbaxes)
```

```python
# Regenerate the first plot for the rush hour timeframe
rush_hr_count = taxi_rides_merged.loc[taxi_rides_merged["lpep_dropoff_hour"].isin([4,5,6,7]), "DOLocationID"].value_counts().reset_index()
rush_hr_count.columns = ["DOLocationID", "count"]
taxi_rides_geo3 = pd.merge(left = geo_data, right = rush_hr_count, how = "left", left_on = "LocationID", right_on = "DOLocationID")


# Plot
fig, ax = plt.subplots(1, figsize=(20, 8))
ax.axis('off')
ax.set_title("Dropoffs by taxi zone (between 4pm and 7pm)", fontdict={'fontsize': '25', 'fontweight': '3'})
taxi_rides_geo3.plot(column="count", ax=ax, cmap = "Blues")

sm = plt.cm.ScalarMappable(norm=plt.Normalize(vmin=taxi_rides_geo3["count"].min(), vmax=taxi_rides_geo3["count"].max()), cmap="Blues")
sm._A = []
cbaxes = fig.add_axes([0.15, 0.25, 0.01, 0.4])
cbar = fig.colorbar(sm, cax=cbaxes)
```
